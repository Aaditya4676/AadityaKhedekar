```
# importing libraries 
import numpy as np
import matplotlib.pyplot as plt
import wave
#from playsound import playsound

#from IPython.display import Audio

np.set_printoptions(threshold=np.inf)
```

```
def dtft(sig,n,beg,s):
  # creating the w array
  w = np.linspace(-np.pi, np.pi, s)  
  # creating the n array
  i = np.linspace(beg,beg + n - 1, n)
  # Simplifying DTFT into X = signal*exp(jw'n)', where w is the vector with omega values, n is the vector with n values and j is iota
  a = np.complex(0,-1)*np.outer(w,i)
  a = np.exp(a)
  X = sig@np.matrix.transpose(a)
  return X
```

```
def convolution(x,h):
  # length of convolution output
  L = len(x) + len(h) + 1

  X = np.zeros(L+1)
  X[0:len(x)] = x
  H = np.zeros(L+1)
  H[0:len(h)] = h
  Y = np.zeros((L+1), dtype = 'float')
  for i in range(L+1):
    for j in range(0,i+1):
      Y[i] += X[j]*H[i-j]
  Y = Y[0:L-2]
  return(Y)
```

Defining Plotting Functions :

```
# Defining a function for plotting a complex signal
def plotsignal(sig,n,beg):
  fig, axs = plt.subplots(2)
  fig.suptitle("Magnitude and Phase plots")
  t = np.linspace(beg, beg + n -1, n)
  print("tt  ",t )
  # print (np.size(t))
  axs[0].stem(t,np.abs(sig), use_line_collection = 'true')
  axs[0].set_xlabel('time (n)')
  axs[0].set_ylabel('Magnitude')
  axs[1].stem(t,np.angle(sig), use_line_collection = 'true')
  axs[1].set_xlabel('time (n)')
  axs[1].set_ylabel('Phase')
  plt.show()

# Defining a function for plotting a signal vs time
def plotsignaltime(sig,n,beg):
  fig, axs = plt.subplots(2)
  fig.suptitle("Magnitude and Phase plots")
  t = np.arange(len(sig))/float(beg)
  # print (np.size(t))
  axs[0].stem(t,np.abs(sig), use_line_collection = 'true')
  axs[0].set_xlabel('time (n)')
  axs[0].set_ylabel('Magnitude')
  axs[1].stem(t,np.angle(sig), use_line_collection = 'true')
  axs[1].set_xlabel('time (n)')
  axs[1].set_ylabel('Phase')
  plt.show()

# Defining a function for plotting a complex signal in terms of real and complex parts
def plotsignalcomp(sig,n,beg):
  fig, axs = plt.subplots(2)
  fig.suptitle("Real and Imaginary plots")
  t = np.linspace(beg, beg + n - 1 , n)
  axs[0].stem(t,np.real(sig), use_line_collection = 'true')
  axs[0].set_xlabel('time (n)')
  axs[0].set_ylabel('Amplitude of real part')
  axs[1].stem(t,np.imag(sig), use_line_collection = 'true')
  axs[1].set_xlabel('time (n)')
  axs[1].set_ylabel('Amplitude of real part')

  plt.show()


# Defining a function for plotting a complex signal in frequency domain
def plotfouriersignal(sig,n,beg,end):
  fig, axs = plt.subplots(2)
  fig.suptitle("Magnitude and Phase plots of the fourier transform")
  t = np.linspace(beg, end, n)
  # print (t)
  axs[0].stem(t,np.abs(sig), use_line_collection = 'true')
  axs[0].set_xlabel('Frequency (w)')
  axs[0].set_ylabel('Magnitude')
  axs[1].stem(t,np.angle(sig), use_line_collection = 'true')
  axs[1].set_xlabel('Frequency (W)')
  axs[1].set_ylabel('Phase')
  plt.show()


# Defining a function for plotting a complex signal
def plotrawsignal(sig,n,sig1,n2,beg):
  fig, axs = plt.subplots(2)
  fig.suptitle("Magnitude and Phase plots")
  t = np.linspace(beg, beg + n - 1, n)
  t1 = np.linspace(beg, beg + n2 - 1, n2)
  # print (np.size(t))
  axs[0].stem(t1,sig1, use_line_collection = 'true')
  axs[0].set_xlabel(' n')
  axs[0].set_ylabel('Input Signal X[n]')
  axs[1].stem(t,sig,  use_line_collection = 'true')
  axs[1].set_xlabel(' n ')
  axs[1].set_ylabel('Convolution output Y[n]')
  plt.show()
```
